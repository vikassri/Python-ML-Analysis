```python
### Implementing the stack in Python ###
class Stack:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return self.items == []
    
    def push(self,data):
        self.items.append(data)
        
    def pop(self):
        return self.items.pop()
        
    def peek(self):
        return self.items[len(self.items) - 1]
    
    def size(self):
        return len(self.items)
    
    def pprint(self):
        print([x for x in self.items])
        
    def reverse(self):
        items = []
        while not self.is_empty():
            items.append(self.pop())
        for item in items:
            self.push(item)
        return self.items
            
s = Stack()             # Initializing the stack
print(s.is_empty())     # check if stack is empty
s.push(8)               # Push 8 into stack s
s.push('dog')           # push 'dog' into stack s
s.push(5)               # push 5 into stack
s.size()                # check the size of stack
s.pprint()              # print the size of stack
s.pop()                 # pop the first value from stack  
s.pprint()              # print the stack
s.reverse()             # reversing the stack 
```

```python
#### find if braces are balanced or not #### 
def matches(open, close):
    opens = "([{"
    closes = ")]}"
    return opens.index(open) == closes.index(close)

def par_checker(par_string):
    s = Stack()
    balanced = True
    index = 0
    while index < len(par_string) and balanced:
        symbol = par_string[index]
        if symbol in "({[":
            s.push(symbol)
        else :
            if s.is_empty() :
                balanced = False
            else :
                top = s.pop()
                if not matches(top, symbol):
                    balanced = False
                
        index += 1
    if balanced and s.is_empty():
        return True
    else :
        return False
    

print(par_checker('{{([][])}}'))
print(par_checker('{{([][)}()}'))
```

```python
##  Decimal to Binary Conversion Using Stack
def base_converter(num,base):
    digits = "0123456789ABCDE"
    s = Stack()
    while num > 0 :
        rem = num%base
        s.push(rem)
        num = num//base

    result = ""
    while not s.is_empty():
        result =  result + str(digits[s.pop()])

    return result

print(base_converter(25,2))
print(base_converter(29,16))
print(base_converter(27,8))
```

```python

####  INFIX to POST/PREFIX ###########

def infix_to_postfix(infix_expr):
    s = Stack()
    postfix_list = []
    token_list = infix_expr.split()
    pre = {}
    pre['('] = 1
    pre['+'] = 2
    pre['-'] = 2
    pre['*'] = 3 
    pre['/'] = 3
    for token in token_list:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
            postfix_list.append(token)
        elif token == '(':
            s.push(token)
        elif token == ')':
            top_token = s.pop()
            while top_token != '(':
                postfix_list.append(top_token)
                top_token = s.pop()
        else :
            while (not s.is_empty()) and (pre[s.peek()] >= pre[token]):
                postfix_list.append(s.pop())
            s.push(token)
            
    while not s.is_empty():
        postfix_list.append(s.pop())
    return ''.join(postfix_list)

print(infix_to_postfix("( A + B ) * C - ( D - E ) * ( F + G )"))
print(infix_to_postfix("( A + B ) * ( C + D )"))
print(infix_to_postfix("( A + B ) * C"))
```

```python

def do_math(op,op1,op2):
    if op == '*':
        return op1*op2
    elif op == '/':
        return op1/op2
    elif op == '+' :
        return op1+op2
    elif op == '-':
        return op1-op2
    
def postfix_eval(post_expr):
    s = Stack()
    token_list  = post_expr.split()
    for token in token_list:
        if token in "0123456789":
            s.push(int(token))
        else :
            op1 = s.pop()
            op2 = s.pop()
            result  = do_math(token,op1,op2)
            s.push(result)
        
    return s.pop()

print(postfix_eval('7 8 + 3 2 + /'))
```

```python
# Queues Implementation

class Queue:
    def __init__(self):
        self.items = []
        
    def enqueue(self,data):
        self.items.insert(0,data)
    
    def dequeue(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
    
    def is_empty(self):
        return self.items == []
    
    def pprint(self):
        print([x for x in self.items])
    
q = Queue()          # Creating the new Queue Instance
q.enqueue(6)         # Inserting 6 into queue
q.enqueue('dog')     # Inserting 'dog' in queue
q.pprint()           # Print the values
q.dequeue()          # Removing the first value
q.pprint()           # Print the value
```

```python
### Rotating potato

def rotating_potato(name_list,num):
    q = Queue()
    for name in name_list:
        q.enqueue(name)
    while q.size() > 1:
        for i in range(num):
            q.enqueue(q.dequeue())
        q.dequeue()
    return q.dequeue()

print(rotating_potato(["Bill", "David", "Susan", "Jane", "Kent","Brad"], 3))
```

```python
# Implementing the Deque with python

class Deque:
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return self.items == []
    
    def add_front(self,data):
        self.items.append(data)
        
    def add_rear(self,data):
        self.items.insert(0,data)
    
    def remove_rear(self):
        return self.items.pop(0)
    
    def remove_front(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
    
    def pprint(self):
        print([x for x in self.items])
    
# palidrome checker
def pal_checker(a_string):
    char_deque = Deque()
    still_equal = True
    for ch in a_string:
        char_deque.add_rear(ch)
    
    while char_deque.size() > 1 and still_equal:
        first = char_deque.remove_front()
        last = char_deque.remove_rear()
        if first != last:
            still_equal = False
            
    
    return still_equal
    
print(pal_checker("radar"))
print(pal_checker("radio"))
```

```python
# circular linklist
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

    def get_data(self):
        return self.data
    
    def get_next(self):
        return self.next
    
    def set_data(self,data):
        self.data = data
    
    def set_next(self,data):
        self.next = data
        

class circularlinkedlist:
    length = 0
    def __init__(self,head = None):
        self.head = head

    def addfirst(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            new_node.set_next(new_node)
        else:
            current = self.head
            while current.get_next() != self.head:
                    current = current.get_next()
            new_node.set_next(self.head)
            current.set_next(new_node)   
            self.head = new_node
        self.length += 1

    def addlast(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            new_node.set_next(new_node)
        else:
            current = self.head
            while current.get_next() != self.head:
                    current = current.get_next()
            new_node.set_next(self.head)
            current.set_next(new_node)   
        self.length += 1        
    
    def dellast(self):
        current = self.head
        while current.get_next() != self.head:
            previous = current
            current = current.get_next()
        previous.set_next(self.head)
        self.length -= 1        
        
    def delfirst(self):
        current = self.head
        while current.get_next() != self.head:
                current = current.get_next()
        current.set_next(self.head.get_next())
        self.head = current
        self.length -= 1            

    def delatpos(self, index):
        if index == 0:
            self.delfirst()
        elif index == self.length:
            self.dellast()
        else :
            current = self.head
            cnt = 0
            previous = None
            while cnt < index:
                previous = current
                current = current.get_next()
                cnt += 1
            previous.set_next(current.get_next())
            self.length -= 1         
        
    def pprint(self):
        current = self.head
        cnt = 0
        lst = []
        while cnt < self.length:
            lst.append(str(current.get_data()))
            current = current.get_next()
            cnt += 1
        print(' => '.join(lst))

    def get_head(self):
        return self.head

# bonus question (How to figureout if this is circular linked list)
def is_circular(head):
    runner = head.get_next()
    current = head
    circular = False
    while (True):
        if (runner == None ) or (runner.get_next() == None):
            return False
        elif  (runner == current ) or (runner.get_next() == current):
            return True
        else :
            runner = runner.get_next()
            current = current.get_next()
        


c = circularlinkedlist()        
c.addfirst(10)
c.addfirst(13)
c.addfirst(14)
c.addfirst(15)
c.addlast(16)
c.pprint()
c.dellast()
c.pprint()
c.delfirst()
c.pprint()
c.delatpos(1)
c.pprint()
node = c.get_head()
print(is_circular(node))
```

```python
# Sorting 

#- Selection Sort

def selection_sort(a_list):
    n = len(a_list)
    for i in range(n-1):
        smaller = i
        for j in range(i,n):
            if a_list[j] < a_list[smaller]:
                smaller = j
                
        (a_list[i],a_list[smaller]) = (a_list[smaller],a_list[i])
        
#- Insertion Sort

def insertion_sort(a_list):
    n = len(a_list)
    for i in range(1, n):
        key = a_list[i]   # remember what was in the ith position
        j = i-1     # look in positions to the left of i
        while j >= 0 and a_list[j] > key:
            a_list[j+1] = a_list[j]
            j -= 1

        a_list[j+1] = key
        
a_list = [89, 45, 85, 81, 77, 94, 22, 79, 92, 91]
selection_sort(a_list)
print("Selection Sort after sorting:  " + str(a_list))
insertion_sort(a_list)
print("Insertion Sort after sorting:  " + str(a_list))
```
